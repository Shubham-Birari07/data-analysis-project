```python
#line plot
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np



```

```python
#line plot
var = [1,2,3,4,5,6,7]
var_1 = [2,3,4,1,5,6,7]

#using matplotlib
plt.plot(var,var_1)
plt.show()

#using seaborn

sns.lineplot(x=var,y=var_1)
```

```python

# Load your local CSV file
df = pd.read_csv("globalAirQuality.csv")
df = df.head(8000)
df
```

```python
sns.lineplot(data=df,x="latitude",y="longitude")

```

```python
sns.barplot(data=df,x="latitude",y="longitude",orient="h")
plt.show()
plt.tight_layout()
plt.show()
```

```python
sns.barplot(data=df,x="latitude",y="longitude",color="red",capsize=5)
plt.show()
plt.show()
plt.show()
plt.tight_layout()
plt.show()
```

```python
sns.displot(df["pm25"],kde=True,rug=True, log_scale=True) #kde is thar wave line,rug is see at bottom of bars that small lines, log_scale it converts bigs into log function
```

```python
#Scatter plot

sns.scatterplot(data=df,x="latitude",y="longitude",hue="pm25",palette="gist_ncar")#hue is the data which is differentiate, palette is for plot design
```

```python
#Heatmaps plot
var = np.linspace(1,10,20).reshape(4,5)
var
```

```python

sns.heatmap(var,annot=True, annot_kws={"fontsize":20,"color":"r"},linewidths=10,linecolor="y",cbar=True)#annot is the numbers inside the box, annot_kes is no. design inside box,linewidths&color are border of boxes, cbar is side-bar if true is appers if false it doesn't appers
```

```python
sns.heatmap(var,vmin=0,vmax=20,cmap="YlGnBu",xticklabels=False,yticklabels=False) #vmin is minimum value of bar vmax is maximum value, cmap is color grading
```

```python
sns.heatmap(var,vmin=0,vmax=20,cmap="YlGnBu")
```

```python
#box plot

sns.set(style="whitegrid")

sns.boxplot(var,x="day",y="total_bill",palette="plasma",hue="sex",showmeans=True,meanprops={"marker":"+","markeredgecolor":"yellow"})
```

```python
#cat plot

sns.catplot(data=var,kind="boxen") #there are type of plot in cat plot use kind= to define which plot meagerly all same given above plots
```

```python
#facet grid in seaborn

graph = sns.FacetGrid(var,col="day",hue="day")
graph.map(sns.scatterplot,"total_bill","tip").add_legend()
plt.show()
```

```python
graph = sns.FacetGrid(var,col="sex",hue="day",palette="summer")
graph.map(plt.bar,"total_bill","tip").add_legend() #use matplotlib
plt.show()
```
